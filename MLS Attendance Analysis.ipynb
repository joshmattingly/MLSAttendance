```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Datasets

Defunct MLS teams (Chivas, Miami, Tampa Bay) were removed prior to import
Stadium capacity is based on the MLS Wikipedia page (https://en.wikipedia.org/wiki/Major_League_Soccer#Teams)

```python
# import the dataset
table = pd.read_excel('Attendance.xlsx')
city_stad = pd.read_excel('CityStadium.xlsx')
dataset = pd.melt(table, id_vars=['Year'], var_name='Club')
dataset = dataset.merge(city_stad[['Club', 'Stadium Capacity']], on='Club', copy=False)
dataset['Attendance / Capacity'] = dataset['value']/dataset['Stadium Capacity']
dataset.head()
```

```python
def change_width(ax, new_value) :
    for patch in ax.patches :
        current_width = patch.get_width()
        diff = current_width - new_value
        patch.set_width(new_value)
        patch.set_x(patch.get_x() + diff * .5)
```

#### Current Stadium Capacity by Club

```python
a4_dims = (14.7, 6.27)
fig, ax = plt.subplots(figsize=a4_dims)
sns.barplot(x='Club', y='Stadium Capacity', data=city_stad, linewidth=14)
change_width(ax, .45)
```

#### Attendance by Year (Men in Red in, well, red)

```python
a4_dims = (14.7, 6.27)
fig, ax = plt.subplots(figsize=a4_dims)
sns.lineplot(x='Year', y='value', hue='Club', data=dataset[dataset['Club'] != 'CHI'])
sns.lineplot(x='Year', y='value', color='red', linewidth=5, alpha=1, data=dataset[dataset['Club'] == 'CHI'])
plt.legend(bbox_to_anchor=(1.2, 1), borderaxespad=0.)
```
